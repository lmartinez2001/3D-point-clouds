```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['axes.spines.top'] = False
plt.rcParams['axes.spines.right'] = False

vals = np.array([[12, 29, 10, 14, 73],
                [18, 8, 4, 4, 5],
                [10, 2, 7, 80, 299],
                [21, 3, 17, 1, 299],
                [3, 251, 77, 299, 299],
                [1, 299, 200, 299, 290],
                [7, 210, 209, 41, 299],
                [1, 7, 22, 247, 299],
                [12, 8, 299, 24, 299],
                [21, 209, 209, 118, 114]]) +1 


vals_fast = np.array([[ 2,  2,  1, 10,  5],
                        [ 2,  1, 14,  1,  5],
                        [ 1,  6,  6,  4,  1],
                        [ 0, 11,  4,  0, 17],
                        [ 2,  5,  4,  4,  9],
                        [ 0, 13, 17,  6, 46],
                        [ 1,  1, 26,  8, 23],
                        [ 0,  5,  2,  7,  6],
                        [ 1,  2,  4,  6,  2],
                        [ 0,  0, 18, 16,  0]]) + 1
```

```python
plt.figure()
plt.errorbar(x=np.arange(5, dtype=int) + 1, y=np.mean(vals, axis=0), yerr=np.std(vals, axis=0)/np.sqrt(10), fmt='o', label="Vanilla RANSAC")
plt.errorbar(x=np.arange(5, dtype=int) + 1, y=np.mean(vals_fast, axis=0), yerr=np.std(vals_fast, axis=0)/np.sqrt(10), fmt='o', label="N-RANSAC")
plt.xlabel("Plane")
plt.ylabel("Number of draws")
plt.yscale("log")
plt.xticks(np.arange(1, 6, dtype=int))
plt.legend(loc="lower right")
plt.tight_layout()
plt.savefig("../ablation_draws.png")
```
